## Person class

In this section you create a **class** to represent a person.
A class is a blueprint for objects. Each `Person` object has:

- a `name` attribute
- an `age` attribute

You then create one `Person` object and print its data.


```python
# Class representing a person
class Person:
    def __init__(self, name: str, age: int, latitude: float, longitude: float):
        self.name = name
        self.age = age
        self.latitude = latitude
        self.longitude = longitude

    def describe(self) -> str:
        # Format coordinates for Google Maps (latitude,longitude with no spaces)
        google_maps_location = f"{self.latitude},{self.longitude}"
        return (
            f"{self.name} is {self.age} years old at location: {google_maps_location} "
            f"(paste into Google Maps)"
        )

    def happy_birthday(self) -> None:
        """Increase the person's age by 1 year."""
        self.age += 1

# Create several Person objects with home locations
person_haaland = Person(name="Haaland", age=67, latitude=59.9139, longitude=10.7522)
person_malte = Person(name="Malte", age=23, latitude=59.9139, longitude=10.7522)
person_kristoffer = Person(name="Kristoffer", age=24, latitude=59.9139, longitude=10.7522)
person_william = Person(name="William", age=23, latitude=59.9139, longitude=10.7522)
person_ibrahim = Person(name="Ibrahim", age=23, latitude=59.9139, longitude=10.7522)
person_daniel = Person(name="Daniel", age=23, latitude=59.9139, longitude=10.7522)

people = [
    person_haaland,
    person_malte,
    person_kristoffer,
    person_william,
    person_ibrahim,
    person_daniel,
]

print("Before birthdays:")
for p in people:
    print(p.describe())

print("\nAfter birthdays (one by one):")
for p in people:
    p.happy_birthday()
    print(p.describe())
```

```python
# Create a list of random Person objects in Bern
import random

names = ["Haaland", "Malte", "Kristoffer", "William", "Ibrahim", "Daniel"]

people_random = []
for _ in range(10):
    name_choice = random.choice(names)
    age_choice = random.randint(20, 63)
    # Bern, Switzerland (approximate coordinates)
    person = Person(
        name=name_choice,
        age=age_choice,
        latitude=46.9480,
        longitude=7.4474,
    )
    people_random.append(person)

print(f"Created {len(people_random)} persons.")
print("Sample person:", people_random[0].describe())
```

## List of European capitals

In this section you create a **list** of European capital cities.
A list stores several values in order. You can loop over the list
and print each capital.


```python
# List of some European capitals
capitals = [
    "Oslo",       # Norway
    "Stockholm",  # Sweden
    "Copenhagen", # Denmark
    "Helsinki",   # Finland
    "Berlin",     # Germany
    "Paris",      # France
    "Madrid",     # Spain
    "Rome",       # Italy
    "Amsterdam",  # Netherlands
]

print("European capitals:")
for capital in capitals:
    print("-", capital)
```

## Adding a new capital to the list

Here you take the existing list of European capitals
and use the `append` method to add a new city: **Vilnius**.
Then you print the updated list.


```python
# Add a new capital to the existing list
capitals.append("Vilnius")

print("Updated European capitals:")
for capital in capitals:
    print("-", capital)
```

## Removing a capital from the list

Now you remove one city from the existing list of capitals.
You use the `remove` method to delete **Madrid** from the list,
then print the updated list.


```python
# Remove a capital from the existing list
if "Madrid" in capitals:
    capitals.remove("Madrid")

print("European capitals after removing Madrid:")
for capital in capitals:
    print("-", capital)
```

## Finding the "happiest" capital

Here you remove the **first** capital from the existing list using `pop(0)`.
You treat this removed city as the one with the **most happiness**, then
print that city and the remaining capitals.


```python
# Remove the first capital and treat it as the "happiest" one
if len(capitals) == 0:
    print("No capitals left in the list. Resetting the list.")
    capitals = [
        "Oslo",       # Norway
        "Stockholm",  # Sweden
        "Copenhagen", # Denmark
        "Helsinki",   # Finland
        "Berlin",     # Germany
        "Paris",      # France
        "Madrid",     # Spain
        "Rome",       # Italy
        "Amsterdam",  # Netherlands
    ]

# At this point, we know there is at least one capital
happiest_capital = capitals.pop(0)
print("Capital with the most happiness:", happiest_capital)

print("\nRemaining European capitals:")
for capital in capitals:
    print("-", capital)
```

## Random persons at different locations in Bern

This cell creates a list of 10 `Person` objects using a **for loop**.
Each person has:

- a randomly chosen name
- a random age between 20 and 63 years
- a **different location within Bern** (latitude and longitude vary slightly for each person)

After creating the list, a sample person is printed.


```python
# Create a list of 10 random persons with different locations in Bern
import random

names = ["Haaland", "Malte", "Kristoffer", "William", "Ibrahim", "Daniel", "Emma", "Sophie", "Lucas", "Oliver"]

# Bern center coordinates
bern_lat = 46.9480
bern_lon = 7.4474

bern_people = []
for _ in range(10):
    name_choice = random.choice(names)
    age_choice = random.randint(20, 63)
    
    # Add small random variation to create different locations within Bern
    # (roughly ±0.01 degrees = ~1 km variation)
    lat_variation = random.uniform(-0.01, 0.01)
    lon_variation = random.uniform(-0.01, 0.01)
    
    person = Person(
        name=name_choice,
        age=age_choice,
        latitude=bern_lat + lat_variation,
        longitude=bern_lon + lon_variation,
    )
    bern_people.append(person)

print(f"Created {len(bern_people)} persons in different locations in Bern.")
print("\nSample person:")
print(bern_people[0].describe())
```
